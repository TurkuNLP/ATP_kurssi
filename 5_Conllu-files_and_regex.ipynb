# Syntaksianalysoidut aineistot

- Juoksevan tekstin sijaan nykyään tekstiaineistot ovat aika usein syntaksimerkittyjä. 

- Yhä yleistyvänä standardina syntaksia kuvataan dependenssipuilla, joissa sanojen roolia ja sanojen välisiä riippuvuuksia kuvataan dependensseillä eli riippuvuuksilla. 

- Riippuvuudet sitovat yhteen pääsanan (head/governor) ja sille alisteisten sanan (dependent)

- Riippuvuuksilla on aina suunta: ne lähtevät pääsanasta

- Riippuvuuksilla on aina joku tyyppi, esim. subjekti tai objekti

- Riippuvuuksia on aina vain yksi per sana

- Koko virkkeen "juuri" on sana, josta kaikki muut sanat ainakin välillisesti lähtevät. Näitä on vain yksi per virke.

- Tavallisesti juuri on finiittiverbi

- Lähtökohtana on, että riippuvuudet eivät saa mennä "ristiin", vaan niitä seuraamalla pitäisi päästä virkkeen juureen.

- Käydään kurkkaamassa pari esimerkkiä Parsebankista: http://bionlp-www.utu.fi/dep_search/ 



- Syntaksitiedot merkitään erilaisilla conll-formaateilla. Näitä on useampia (conllu, conll09, jne.), mutta pääidea on, että teksti on sana / rivi -muodossa, ja rivillä tiedot sanan paikasta virkkeessä, perusmuodosta, sanaluokasta ja syntaksifunktiosta ovat omilla sarakkeillaan.

Esim. alla

1. sarake kertoo sanan järjestysnumeron virkkeessä, ja toimii myös sen id:nä

2. sana

3. perusmuoto eli lemma

4. sanaluokka

6. morfologiset tiedot

7. sanan pääsanan id, eli sana, josta tähän sanaan tulee riippuvuus. 0 tarkoittaa, että sana on virkkeen juuri.

8. riippuvuuden tyyppi

1       Lue     lukea   VERB    _       Mood=Imp|Number=Sing|Person=2|VerbForm=Fin|Voice=Act    0       root    _       _

2       lisää   lisää   ADV     _       _       1       advmod  _       _


# Syntaksianalysoitujen aineistojen käsittelyä

- Polulta mavela@nlpedu1:~/regex-hakemisto$ löytyy  pb_head.conllu.gz-niminen tiedosto, joka sisältää vajaa puoli miljoonaa sanaa internetistä imuroitua suomea. Kopioi se omaaan kotihakemistoosi.

- Tiedosto on pakattu, ja siksi muotoa .gz. Tiedostoa voi katsoa käskyllä zcat, joka myös purkaa sen. Myös zless toimii. 

- esim. zcat  pb_head.conllu.gz

- Tiedosto on conllu-formaatissa, jossa jokainen sana on omalla rivillään. Lisäksi ns. kommenttirivillä on erilaisia tietoja virkkeestä. 

# Harjoitus 1

- Kuinka monta juoksevaa sanaa (sanetta) tässä tiedostossa on?

- Kuinka monta erilaista sanetta tiedostossa on? (välimerkit voi jättää mukaan)

- Kuinka monta virkettä tiedostossa on?

- Mitkä ovat tiedoston viisi yleisintä sanaa?

- Entä viisi yleisintä sanan perusmuotoa?

- Entä viisi yleisintä verbiä? (Näiden etsimiseen riittää, kun etsii egrepillä riviä, jossa on verbin sanaluokkatunniste VERB.)

- Entä viisi yleisintä nominaalisubjektia? (nsubj)

- Kuinka monta sanaa mainitaan vain kerran?

- Ainoastaan kerran mainituissa sanoissa on mukana paljon roskaa, numeroita, analysointivirheitä ja symboleita. Miten näitä saisi pois?

- Saatko laskettua, kuinka monta vain kirjaimista koostuvaa, vain kerran mainittavaa sanaa aineistossa on? Entä vain kerran mainittavia lemmoja?

- Vieraskielisiä sanoja merkitään tagillä Foreign. Mitkä ovat yleisimmät? Millä yksinkertaisella tavalla tätä listaa saa hiukan putsattua?

Vinkkejä:

- egrep -v:llä saat pois ne rivit, joita et halua laskuihin mukaan. 

- Kommentit alkavat aina risuaidalla #. Näihin osuu käsky egrep "^#" . ^ merkitsi rivin alkua.

- Säännöllinen lause tyhjiin riveihin oli "^$", esim. egrep "^$" .

- Virkkeissä sanoilla on järjestysnumero, ja virkkeen ensimmäinen sana on aina 1.

- Rivejä, jotka alkavat numerolla 1 (muttei 11), voi etsiä käskyllä "egrep "^1[[:space:]]"

- cut f 2 printtaa ainoastaan sarakkeen 2, jossa on sane. Sarakkeessa 3 on perusmuoto.

- Kuten aiemmin huomattiin, sort uniq -putken tuottamissa frekvenssilistoissa on ihmeellisiä tyhjiä välejä esimerkiksi heti rivin alussa. Nämä saa pois esim. näin: zcat pb_head.conllu.gz | cut -f2 | sort | uniq -ci | sort -n | perl -pe 's/^ +//g' . Eli perl ... muuttaa rivin alussa (^) olevat 1 tai useammat välit ( +) pois (//). 

- Frekvenssilistasta saa printattua pelkästään saneet käskyllä perl -pe 's/^ +//g' | cut -f 2 -d ' ' 

- Lähtökohtaisesti cut -f käyttää printtaamiensa sarakkeiden erottamiseen tabulaattoria. Jos sarakkeiden välillä on jotain muuta, sen voi ilmoittaa vivulla -d. Esim. -d ' ' ilmoittaa, että sarakkeiden välissä onkin väli.


- Jos haluaa grepata pelkästään kirjaimia sisältäviä "saneita", kannattaa käyttää komplementtimerkkiä. Esim. [^0-9] etsii kaikkia PAITSI numeroita. Saman voi tehdä kirjaimille [^a-zöäåA-ZÖÄÅ].
