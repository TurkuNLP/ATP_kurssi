b) Make a script that downloads a Gutenberg book and prints a clean word frequency list (to standard output or to a file)
* give the http address to the book file as an argument (use double quotes to wrap the http address), not through a pipe!
* Note that `wget -O` allows you to name the downloaded file as you like

Script:
```
#!/bin/bash
wget -O temp-book.txt $1
cat temp-book.txt |
  tr '[:upper:]' '[:lower:]' |
  tr ' ' '\n' |
  egrep '[a-zA-Z]' |
  sort | uniq -c | sort -nr |
  head -10
```
Usage:
```
./script.sh https://gutenberg.org/cache/epub/77088/pg77088.txt
```

N.B. Good practice would be to use a backslash in multiline commands like this:
```
#!/bin/bash
wget -O temp-book.txt $1
cat temp-book.txt | \
  tr '[:upper:]' '[:lower:]' | \
  tr ' ' '\n' | \
  egrep '[a-zA-Z]' | \
  sort | uniq -c | sort -nr | \
  head -10
```
