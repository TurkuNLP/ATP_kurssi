# Tiedostojen järjestämisestä ja frekvenssilistoista


# sort 

Sort järjestää tiedoston aakkosjärjestykseen. HUOM: less putken perässä helpottaa luettavuutta...

-- cat data-per-line.txt | sort | less

sort -n toimii numeroille:

-- cat dates.txt | sort -n | less


# uniq

uniq filtteröi peräkkäisiä, samoja rivejä. Esim. seuraavassa sort järjestää rivit, ja uniq printtaa jokaista samanlaista vain yhden:

--  cat data-per-line.txt | sort | uniq | less

Uniqilla on monta hyödyllistä vipua:

-- -d printtaa vain duplikaatit, ja jokaista vain yhden
-- -D printtaa vain duplikaatit, mutta kaikki ne
-- -c laskee, kuinka monta samoja peräkkäisiä rivejä  on


# Frekvenssilaskuri sort | uniq -c | sort -n -yhdistelmällä

uniq -c laskee perättäisten duplikaattien määrän. Kun tämän yhdistää sort-komentoon, saadaan rivien frekvenssilista!

--  cat data-per-line.txt | sort | uniq -c

sort yksinään ei vielä osaa listata numerofrekvenssejä oikein (aina)

--  cat data-per-line.txt | sort | uniq -c | sort

--  cat data-per-line.txt | sort | uniq -c | sort -n

sort -n järjestää listan numeroiden mukaan, suurimmasta pienimpään

--  cat data-per-line.txt | sort | uniq | sort -n

esim cat t.txt | sort > sorted_t.txt

mitä sort -r tekee? 

esim cat t.txt | sort > sorted_t.txt

# Uusia käskyjä: tr, gzip ja shuf

tr muuttaa merkkijonon kirjaimet isoista pieniksi tai toisin päin:

- tr '[:upper:]' '[:lower:]'

- cat tiedosto.txt | tr '[:upper:]' '[:lower:]'

gzip pakkaa tiedoston, jolloin se vie vähemmän tilaa

- cat tiedosto.txt | gzip > tiedosto.txt.gz

catin versio pakatuille tiedostoille on zcat

- zcat tiedosto.txt.gz 

less on zcatin jälkeen kätevä, toki on myös käsky zless

- zcat tiedosto.txt.gz | less

shuf sekoittaa inputin rivit. tällä saa aineiston sekaisin esimerkiksi random samplea varten.

- zcat tiedosto.txt.gz  | shuf
