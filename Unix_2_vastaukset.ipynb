## Tehtävä 3

- Koita head -komentoa määrittämällä vivun kanssa, kuinka monta riviä haluat nähdä.

head -10 tviitteja1.txt


- Entä miltä näyttää tiedoston viimeinen tviitti?

tail -1 tviitteja1.txt

(Itse asiassa tämä viimeinen rivi on tyhjä. Viimeistä edellinen tviitti on "Viranomainen: "Maassa maan tavalla" on keskeinen osa turvapaikanhakijoiden saamaa koulutusta".


- Kuinka monta riviä tviittejä on? Entä kuinka monta sanaa aineistossa on?

wc -l tviitteja1.txt

wc -w tviitteja1.txt

- Kuinka monessa tviitissä mainitaan sana Suomi? Entä suomi?

egrep "Suomi" tviitteja1.txt 

egrep "suomi" tviitteja1.txt

- Ohjaa omaan tiedostoonsa kaikki tviitit, joissa mainitaan sana suomi. Laske, kuinka monta tviittiä tässä tiedostossa on. Tämän jälkeen ohjaa tähän tiedostoon myös kaikki ne tviitit, joissa mainitaan "Suomi". Laske sitten, kuinka monta näitä tuli yhteensä.

egrep "suomi" tviitteja1.txt > tiedosto.txt

wc -l tiedosto.txt

egrep "Suomi" tviitteja1.txt >> tiedosto.txt

wc -l tiedosto.txt


## Tehtävä 4

1. Kopioi kansiosta mavela/dataa tiedosto data.txt omaan kansioosi -- esim: cp /home/mavela/dataa/data.txt .

2. Selaile tiedostoa data.txt eri komennoilla. Miksi cat on tähän epäkätevä, ja esim. head kätevä? Mitä tiedosto sisältää? Kuinka monta sanaa siinä on? Entä rivejä?
    -- sanoihin:
        -- tapa 1: wc -w data.txt
        -- tapa 2: cat data.txt | wc -w
    -- riveihin:
        -- tapa 1: wc -l data.txt
        -- tapa 2: cat data.txt | wc -l
   
3. Kuinka monella rivillä "ihminen" mainitaan? Entä kuinka monella rivillä ei ole merkkijonoa "ihminen"?
        -- cat data.txt | egrep "ihminen" | wc -l
        -- cat data.txt | egrep -v "ihminen" | wc -l

4. egrepillä on monia näppäriä vipuja. Koita ymmärtää man-sivujen, internetin ja yrittämisen avulla, mitä vivut -n, -c, -i ja -w tekevät.
    --  -n lisää rivinumerot
    -- -i ei tee eroja isojen ja pienten kirjaimien välillä (ignore case)
    -- -c laskee osumat
    -- -w osuu vain kokonaisiin sanoihin. esim. egrep "on" palauttaisi sanan "jono", mutta egrep -w "on" ei. 

Huomaathan, että vipuja voi myös yhdistellä. Mitä siis tekisi käsky egrep -vi? Entä egrep -ciw?

    -- egrep -vi palauttaa ne rivit, joilla ei ole osumia, eikä ota huomioon isojen ja pienten kirjainten eroa
    -- egrep -ciw laskee, kuinka monella rivilla osuma on omana sananaan sekä isoilla että pienillä kirjaimilla
    
5. Kuinka monella rivillä tiedostossa esiintyy merkkijono "on"? Entä sana "on"? Entä kuinka monessa ei esiinny sanaa "on"? Entä, jos laskuihin ottaa mukaan isot ja pienet kirjaimet?

    -- cat data.txt | egrep "on" | wc -l
    -- cat data.txt | egrep -w "on" | wc -l
    -- cat data.txt | egrep -vw "on" | wc -l
    -- cat data.txt | egrep -vwi "on" | wc -l
    
6. Keskustelujen aiheiden piti liittyä yhteiskuntaan, mutta mukana on aika paljon uskontoa. Kuinka monella rivillä mainitaan Jeesus jossain muodossa? Ota pois kaikki nämä rivit, ja ohjaa tuotos tiedostoon putsattu.txt. Kuinka monta riviä tässä putsatussa versiossa on?

    -- cat data.txt | egrep -vi "jeesus" > putsattu.txt
    -- cat putsattu.txt
