# Säännölliset lausekkeet

- Merkkijonokieli, joka voi vastata tai olla vastaamatta toista merkkijonoa

    - Näppärä etsimiseen, korvaamiseen, monen tiedoston samanaikaiseen ajamiseen...
    
    - Huom! Saattavat vähän vaihdella eri kielissä, eil esim. bashin eli komentorivin ja perl-komennon välillä voi olla joitain eroja. Ei välitetä siitä.
    
    - Erilaisia operaattoreita:
    
    - ^ --> merkkijonon / rivin alku
    
    -  $ --> merkkijonon / rivin loppu
    
    - ^$ --> tyhjä rivi (kätevä eritoten grep -v:n kanssa!)
    
    - . --> mikä tahansa merkki
    
    - | --> vaihtoehto
    
    - \* --> match edellinen merkki 0 kertaa tai useammin
    
    - \+ --> match edellinen merkki 1 kerran tai useammin
    
    - ? --> match edellinen merkki 0 tai 1 kertaa
    
    - [] --> ryhmä, esim. [A-ZÅÄÖa-zåäö] --> kaikki kirjaimet, [0-9] --> kaikki numerot
    
    - toki sama asia voidaan ilmaista myös eri tavoin, esim [abc] on sama kuin (a|b|c)
    
    
    - a{m}  m kertaa a
    
    - a{m,}	ainakin m kertaa a
    
    - Näitä voi yhdistää niin, että esim. 
    
            -- ".*" --> mikä tahansa merkki 0 kertaa tai useammin 
            
            -- ".?" --> mikä tahansa merkki 0 tai 1 kertaa
            
            -- ".+" --> mikä tahansa merkki 1 kertaa tai useammin
            
            --"a+" --> a-kirjain 1 kertaa tai useammin
            
            -- "a.*" --> a-kirjain, mikä tahansa merkki 1 kertaa tai useammin
            
            -- "a?.$" --> a-kirjain 0 tai 1 kertaa, mikä tahansa merkki, rivinloppu
               
- HUOM: jos haluaakin etsiä kirjaimellisesti jotain yleensä säännöllisiin lausekkeisiin kuuluvaa merkkiä, pitää merkin eteen laittaa \. Esim. \^ etsii merkkiä, ei rivin alkua.

- Säännöllisillä lausekkeilla voi tehdä vaikka mitä. Esimerkiksi komento ls .\*py listaa kaikki py-loppuiset tiedostot.

- Esimerkkejä säännöllisistä lausekkeista egrepin avulla:

    -- kaikki rivinalkuiset merkkijonot "on": (huom: tähänkin voisi näppärästi yhdistellä egrepin vipuja!)
    
        -- egrep "^on"
        
    -- kaikki rivinloppuiset merkkijonot "on":
    
        -- egrep "on$"
        
    -- kaikki rivit, joissa on kirjain p:
   
        -- egrep "p"
        
    -- oletetaan, että tiedostossa on seuraavat sanat omilla riveillään: 
            pöö
            ppöö
            pppöö
            öö
            pööp

        -- egrep "p.$" palauttaisi rivit, jossa on p, yksi mikä tahansa merkki ja rivinloppu. Näitä ei ole.
        
        -- egrep "p.+$" palauttaisi rivit, joissa on p, yksi mikä tahansa merkki (.) kerran tai useammin (+) ja rivinloppu:
        
            pöö
            ppöö
            pppöö
    
        -- egrep "pööp?$" palauttaisi rivit, joissa on pöö, 0 tai 1 kertaa p ja rivinloppu:
        
            pöö
            ppöö
            pppöö

       -- egrep "^ö" palauttaisi ö-alkuiset rivit:
       
           öö
           
        -- egrep "^p?öö" palauttaisi kaikki rivin, joiden alussa on 0 tai 1 p-kirjainta sekä öö:
        
            pöö
            öö

-- Esimerkkejä vaihtoehdoista putken (|) avulla

        -- egrep "ja|mutta" palauttaa rivit, joissa jompi kumpi näistä
        
    -- Sulkeita voi käyttää ryhmittelyyn:
        
        -- egrep "h(a|ä)n" palauttaa rivit, joissa on joko han tai hän
        
    -- Sulkeiden jälkeen voi käyttää muita säännöllisiä lausekkeita, jotka osuvat sulkeiden sisältään. 

           -- egrep "m(in)*ä"

                minä
                mä

- Merkkiryhmiin voi viitata hakasulkeiden avulla

        -- egrep "[0-9]" palauttaa kaikki rivit, joissa on joku numero
        
        -- egrep "[A-ZÅÄÖa-zÅÄÖ]" palauttaa kaikki rivit, joissa on joku kirjain

      -- Näihinkin voi yhdistää muita:
      
          -- egrep "^[0-9]+[A-ZÅÄÖ]" palauttaa rivin, jotka alkavat yhdella numerolla, jonka jälkeen on iso kirjain.
