# Tehtävä 5

- Tähän tehtävään käytetään jo aiemmin käytettyä data-per-line.txt -tiedostoa. Se löytyy myös kansiosta /home/mavela/dataa

- Koita ensin sort-komentoa tiedoston järjestämiseen. Nyt muuten tail-komento on kätevä outputin selailuun, sillä pelkkä less putken lopussa näyttää ainoastaan tiedoston alun...

cat data-per-line.txt | sort | less

cat data-per-line.txt | sort | tail

- Mitä sortin -r -vipu tekee tässä komentoyhdistelmässä?

- Tässä vaiheessa on hyödyllistä laskea, kuinka monta riviä eli sanaa tiedostossa on?

cat data-per-line.txt | wc -l
100000

- Laske uniqilla, kuinka monta uniikkia (=erilaista) riviä eli sanaa tiedostossa on? 

cat data-per-line.txt | sort | uniq | wc -l

29512

-Kuinka monta riviä on enemmän kuin yksi?

cat data-per-line.txt | sort | uniq -d | wc -l

8263

- Mikä on tiedoston aakkosjärjestysessä viimeinen sana?

cat data-per-line.txt | sort | uniq | tail -1

zto

- Mitkä ovat tiedoston yleisimmät sanat? 

cat data-per-line.txt | sort | uniq -c | sort -rn | less

- Printtaa vielä tiedostoon 10 yleisintä sanaa 

cat data-per-line.txt | sort | uniq -c | sort -rn | head -10 > frek.txt

- Lisää tiedostoon sitten vielä 10 harvinaisinta sanaa

cat data-per-line.txt | sort | uniq -c | sort -rn | tail -10 >> frek.txt

- Laske egrepin avulla, kuinka sanaa esiintyy vain kerran. HUOM: frekvenssilistassa 1-numeron molemmin puolin on whitespace eli tyhjä väli

 cat data-per-line.txt | sort | uniq -c | sort -rn | egrep " 1 " | wc -l
 
21249


- Tee vielä frekvenssilista, josta vain kerran esiintyvät sanat on pois, ja printtaa se tiedostoon


cat data-per-line.txt | sort | uniq -c | sort -rn | egrep -v " 1 " > frek2.txt
