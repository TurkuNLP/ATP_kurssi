### Using `tr` to normalize

`tr` can be used to **normalize data**, i.e. changing or removing tokens so that same words are recognized as the same, i.e. that e.g.,
  
    cat
    Cat
    cat.
    CAT
    cat,

are recognized as the same word (cat).

With `tr` we can

* **normalize letters** from upper case to lower case (i.e. replace any upper case letter with a lower case letter):

  `tr '[:upper:]' '[:lower:]'`
  
* **delete numbers** (i.e. replace any number `[0-9]` with a whitespace):

  `tr '[0-9]' ' '`
* **delete punctuation** (i.e. replace any punctuation `[:punct:]` with a whitespace):

  `tr '[:punct:]' ' '`


Let's practice these. In the cleaned.txt.gz file

10. normalize the text,
11. delete the numbers and
12. delete the punctuation.


```python
#Start by checking the file contents

! zcat cleaned.txt.gz | head -10
```

```python
#10. Normalize - replace upper case with lower case.

! zcat cleaned.txt.gz  | tr '[:upper:]' '[:lower:]' | head -20
```

```python
#11. Delete numbers.

! zcat cleaned.txt.gz  | tr '[0-9]' ' ' | head -20
```

```python
#12. Delete punctuation.

! zcat cleaned.txt.gz  | tr '[:punct:]' ' ' | head -20
```

**Note** that the `tr` command **replaces with white space**, and if you have e.g. a year with 4 digits, you get 4 extra white spaces. This leads to empty lines in e.g. frequency lists, but the empty lines can be removed by grepping them.

**Q: How would you check the number of columns in a file?**

A: There is no direct way to find out, but if you know the separator (e.g. tab or '\t' in a tab separated file, or ',' or ';' in a comma separated file), you can count how many times the separator exists in the file and add one to get the number of columns (this is like counting the spaces between your own fingers, which is four, and adding one to get the number of fingers). This may not work if the separator exists within the column contents as well, but that would be bad practice from the original creator of the file.

**Q: What is the difference between single and double quotes in Bash?**

A: This is a complicated thing. In brief, both denote strings in Bash, or something that is "just words" and not commands, but with double quotes you can insert something "extra" within the string, e.g. a variable or even a command. From the Bash manual (here: https://www.gnu.org/software/bash/manual):
* Enclosing characters in single quotes (') preserves the literal value of each character within the quotes.
* Enclosing characters in double quotes (") preserves the literal value of all characters within the quotes, with the exception of $, `, \\, [...] The special parameters * and @ have special meaning when in double quotes (see Shell Parameter Expansion).

The differences in command behaviour between single and double quote use is not always very easy to uderstand, especially with egrep where you use regular expressions. But if you want to make absolutely sure that e.g. an echo gives you the literal string, use single quotes. If you want to echo a string including a variable, use double quotes. An example below:

```python
# single quotes: the dollar sign and wavy brackets are interpreted as a dollar sign and wavy brackets
! x='yes' ; echo 'thi$ i$ an example of a variable use within a string: ${x}'
! echo '-------------------'
# double quotes: the dollar sign and wavy brackets are interpreted as "print here whatever variable x contains"
! x='yes' ; echo "thi$ i$ an example of a variable use within a string: ${x}"
```

* All this becomes super difficult when trying things out with egrep.
* **Ignore this explanation if you feel like you are already lost :)** Just prefer single quotes with egrep but keep checking that it does what you want.
* Bash shell and egrep both make their own interpretations of a command but Bash shell does its interpretation first and then gives the result to egrep. So a command like ` egrep "i\$" ` is first interpreted by Bash shell and since we used double quotes and escaped the dollar sign, Bash thinks we meant literal "i\$", which is then given to egrep as "i\$" which egrep interprets as "i at the end of line" and not as literal "i and a dollar sign". If we want egrep to understand that we want to look for the literal dollar sign, we must use single quotes like this `egrep 'i\$'` so that Bash shell does not make its own interpretations and just hands the command over to grep as "i\\$". Then egrep sees that we want to use the dollar sign in the literal sense.



```python
# Examples of single and double quotes with egrep
# (note to self: this is easier to show on the university server because we can see the colors)

# both of these commands do the same thing: they match the last 'i' at the end of line and not the literal 'i$'
! echo 'thi$ i$ an example iiiiiiii' | egrep 'i$'  # Bash shell makes no interpretations of its own because of single quotes
! echo '-----'
! echo 'thi$ i$ an example iiiiiiii' | egrep "i$"  # Bash shell would make interpretations if there was anything to interpret; a single dollar sign means nothing to Bash

# if you really want to search for a literal dollar sign, use single quotes and a backslash: '\$'
! echo '-----'
! echo 'thi$ i$ an example' | egrep "i\$"  # Bash will interpret the backslash, so egrep looks for 'i at the end of a line'
! echo '-----'
! echo 'thi$ i$ an example' | egrep 'i\$'  # Bash will not interpret the backslash, so egrep looks for literal i and a dollar sign
```

* All in all, it may be safer to use single quotes, but always check that the behavior is what you wanted.

```python
! echo 'this is a dollar: 1$' | egrep '1\$'
```

## IV Regular expressions

Above, we saw that **regular expressions** can be used to match a larger group of strings
* `[:punct:]`
* `[:upper:] [:lower:]`
* `[0-9]`

**Note1**: regexes can vary between languages

**Note2**: the letters ****, ****, **** are not recognized e.g. by `[:upper:] [:lower:]`

Some useful **operators**
* `^` beginning of line
* `$` end of line
* `^$` empty line (beginning + end without anything inbetween)
* `|` alternative, e.g., `"cat|dog"`
* `[]` group, e.g.`[A-Za-z]`, `[0-9]`, `[abc]` ***any** of the characters*
* `()` group to form a whole, e.g. `(abc)|(def)`
* The same thing can be expressed in many ways, e.g. `[abc]` is the same as `"a|b|c"`

**NOTE**: if you want to search for the literal meaning of a regular expression, you need to **escape** it with `\`

e.g. `egrep '\$||'`

These (and more) are listed also here: https://www.guru99.com/linux-regular-expressions.html

### A couple more useful operators

* `.` any character
* `*` 0 times or more (i.e. present or absent)
* `+ ` 1 time or more
* `?`  0 or 1 time

Operators can also be **combined**
- `.* ` any character 0 time or more
- `.?` any character 0 or one time
- `.+` any character 1 or more times
- `a+ ` (the letter) *a* 1 or more times
- `a.*` (the letter) *a* followed by any character, 1 or more times
- `a?.$` (the letter) *a* 0 or 1 time followed by any character and line end
