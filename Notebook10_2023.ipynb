# **For loops**

*   with for loops we can execute the same command on several items with just one command


> e.g., in


> ```
for f in cat dog mouse ; do echo $f ; done




> we execute the command `echo` on all the strings `cat` `dog` `mouse` with just one command, instead of running the commands


> `echo cat`


> `echo dog`


> `echo mouse`


> separately.


*   the item on which the command is executed in a for loop can be a file
   

*   this means that we can **execute the same command to multiple files** with one command
*   this is practical, since with for loops we can also execute **scripts** to multiple files with just one command


*   for for loops we need **regular expressions**

**In practise**


*   for loop begins with

>  `for f`



where `f` stands for the variable on which the command is executed (you can think of it as 'file': 'for file in the folder')


*   then the folder with the files


> all files in the folder


> `for f in folder/*`


> files ending with csv


> `for f in folder/*csv`


> files starting with myfi



> `for f in folder/myfi*`



*   OR if you are in the folder already


> all files in the folder


> `for f in *`


> all files ending with .txt


> `for f in *.txt`


*   commands that are executed for each file are put between `;`
*   the command starts with `do`
*   the `$f` refers to the file the command is executed on
*   the loop ends with `done`



> `for f in folder/* ; do echo $f ; done`

*   there can be anything between `;` (pipes, scripts, ...)



> `for f in folder/* ; do cat $f | egrep "snow" ; done`


> `for f in folder/* ; do cat $f | ./script.sh ; done`


**Directing output in a file**



*   the output can be directed into a file



> in the end (all output in **one file**)



> `for f in folder/* ; do cat $f | egrep "snow" ; done > output.txt`




> within the command (each output in **separate files**)



> `for f in folder/* ; do cat $f | ./script.sh > $f-output.txt ; done`


**Several commands with for loop**

*   use several `;` if you wish to execute several commands
*   e.g., here print file names and lines with "snow" in the files



> `for f in folder/* ; do echo $f ; cat $ | egrep "snow" ; done`


**Examples**

*   print **file names and word count**

```
for f in folder/*txt ; do echo $f ; cat $f | wc -w ; done

```

*   print **word count only**

```
for f in folder/*txt ; do cat $f | wc -w ; done

```

*   print **number of file names only**

```
for f in folder/*txt ; do echo $f | wc -w ; done

```

*   print **file names only**

```
for f in folder/*txt ; do echo $f ; done

```

**NOTE**



*   back reference also uses `$`, but works as usual within a for loop



```
for f in *.txt ; do echo $f ; cat $f | perl -pe 's/([[:punct:]])/\n$1/g' ; done
```




*   for all files ending in `.txt`


1.   print the file name
2.   print the file and replace all punctuations with a new line and the punctuation in question (each punctuation is put on a new line)


*   for loops can be used for looping also other items than just files, and the item after `for` (e.g., above: `for f`) doesn't have to be `f`; remember though that it has to match the reference next to `$` (so if you use `for n` make sure you use `$n`)



```
for n in cat dog mouse ; do echo $n ; done
```
