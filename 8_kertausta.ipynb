# Tiedostojen käsittelyä

- grep ja egrep: mätsäävien rivien etsiminen

- egrep -v printtaa rivit, jolla ei ole haluttua asiaa

    -- cat t.txt | egrep -v "koira" # printtaa rivit, jolla ei ole sanaa "koira"
    
 - egrep -c myös laskee rivit
 
 - egrep -w osuu ainoastaan sanoihin, eikä merkkijonoihin jotka sisältävät haetun merkkijonon 

- wc: laskee sanat (-w) ja rivit (-l)

- head, tail: printtaa vain tiedoston alku tai loppu

- putkilla voi yhdistää eri käskyjä:

    -- cat t.txt | egrep "koira" | wc -l # printtaa t.txt-tiedoston, valitsee niistä ainoastaan rivit, joilla on sana "koira", ja laskee nämä rivit
    
    - Komentirivin tuotoksen voi ohjata tiedostoon joko käskyllä > (luo uuden tai uudelleenkirjoittaa olemassaolevan tiedoston) tai >> (lisää olemassaolevaan tiedostoon)

# Merkkijonon muuttaminen toiseksi ja vähän säännöllisiä lausekkeita

- cat t.txt | perl -pe 's/VANHA/UUSI/g' # muuttaa merkkijonon VANHA merkkijonoksi UUSI

- tällä voi myös deletoida esim. pisteitä tai muita välimerkkejä ja erotella sanat omille riveilleen

    - cat t.txt | perl -pe 's/\\.///g' # poistaa pisteet
    
    - cat t.txt | perl -pe 's/ /\n/g' # korvaa välilyönnit rivinvaihdoilla --> laittaa sanat omille riveilleen. Tämän voi hyvin yhdistää yllä esiteltyyn sort-uniq-sort -yhdistelmään, joka laskee erilaisten rivien frekvenssit. Näin saa näppärästi laskettua tiedoston yleisimmät sanat ja niiden frekvenssit.
    
 - toinen hyödyllinen käsky on tr '[:upper:]' '[:lower:]' # muuttaa isot kirjaimet pieniksi
 
 
 - Säännöllisillä lausekkeilla pystyy kätevästi prosessoimaan tekstiä ilman, että kaikkia mahdollisuuksia täytyy kirjoittaa auki.
 
     - Hyödyllisiä säännöllisiä lausekkeita tässä kohtaa on 
     
         --- \n --> rivinvaihto
         
         --- ^$ --> tyhjä rivi
         
         --- [[:space:]] --> monenlaiset tyhjät merkit
         
         --- + --> yksi tai useampi merkki, esim n+ --> yksi tai useampi n-kirjain
         
         --- | --> tai, esim. grep "koira|kissa"--> etsi rivit, joissa joko koira tai kissa

# Pakatut tiedostot, cut -f, syntaksianalysoidut aineistot ja vähän lisää säännöllisiä lausekkeita

- Syntaksijäsennetyt aineistot tulevat conll-muodossa, jossa jokainen sana on omalla rivillään. Ainoastaan tietyn sarakkeen, esimerkiksi saneen perusmuodon, saa printattua käskyllä cut -f (sarakkeen numero). 

    -- esim cat t.conllu | cut f -3
    
 - Syntaksijäsennetyt aineistot ovat helposti isoja, joten ne kannattaa pakata. Tällöin ne loppuvat (esimerkiksi) muotoon -gz. Nämä saa purettua esim. käskyllä zcat
 
     -- esim. zcat t.conllu.gz | less
     
     
 - Säännöllisiä lausekkeita, jotka ovat hyödyllisiä syntaksijäsennettyjen tiedostojen kanssa työskentelyyn
 
     -- ^ --> rivin alku
     
     -- $ --> rivin loppu
     
     -- ^$ --> tyhjä rivi
     
     -- [] -hakasulkeiden sisään voi laittaa merkkiryhmiä, joista johonkin haun pitää osua
     
         --- [0-9] mikä tahansa numero
         
         -- [a-zöäåA_ZÖÄÅ] mikä tahansa kirjain

# Skriptit

- Komentirivitkäskyt voi tallentaa skriptiin, jolloin sitä ei tarvitse aina kirjoittaa uudestaan. Jos skriptille syöttää tiedoston komentorivillä, samaa skriptiä voi näppärästi hyödyntää usemman tiedoston analyysiin.

- Skriptit loppuvat muotoon .sh, esim. skripti.sh. Ne ajetaan käskyllä ./skripti.sh

- Skriptejä voi kirjoittaa esim. tekstieditorilla, kuten gedit.

- Skripteihin voi periaatteessa kirjoittaa mitä vaan. Niiden pitää kuitenkin alkaa käskyllä #!/bin/bash

- Skriptejä voi näppärästi ajaa monelle tiedostolle yhtä aikaa säännöllisten lausekkeiden avulla

    - esim. cat *txt | ./kukkuu.sh 
    
    - esim. zcat *gz | ./haa.sh

----

\#!/bin/bash

cat t.txt | egrep "on" | wc -l

----------

- Lisäksi skripteille pitää erikseen antaa suoritusoikeudet, jotta ne toimivat. Tämä onnistuu esim. käskyllä chmod a+x 

- Tiedostojen oikeuksia pääsee katsomaan käskyllä ls -la

- Skripteihin voi yhdistää käskyjä komentiriviltä

    -- cat t.txt | ./skripti.sh > tulokset.txt # printtaa tiedoston t.txt, ajaa skriptin, ja ohjaa tuotoksen tulokset.txt-nimiseen tiedostoon
    
 
 # Tehtävä
 
 - Tee skripti, jolle annetaan komentiriviltä tiedosto, josta skripti laskee yleisimmät juoksevat saneet, ilman välilyöntejä. Miten skripti voidaan ajaa useammalle tiedostolle yhtä aikaa?

```bash
# Ratkaisu
# muista antaa skriptille suoritusoikeudet!

---------------
#!/bin/bash

egrep -v "^$|^#" | egrep -v "punct" | cut -f 2 | sort | uniq -ci | sort -rn

--------------

# tämä voitaisiin ajaa käskyllä zcat pb.conllu.gz | ./skripti.sh > output.txt, tai useammalle saman hakemiston gz-loppuiselle tiedostolle käskyllä zcat *gz | ./skripti.sh > output.txt
```

# For loopit

- for loop suorittaa käskyn jokaiselle annetulle tiedostolle erikseen. 

    - for tiedosto in hakemisto/*txt ; do echo $f ; done # jokaiselle txt-loppuiselle tiedostolle hakemistossa, tee käsky echo tiedosto
    
- puolipiste ; erottaa eri komentoja toisistaan loopin sisällä, tällöin käskyt tehdään erikseen

    - for tiedosto in hakemisto/*txt ; do echo $f ; cat $ ; done # suorita ensin echo tiedosto, sitten cat tiedosto
    
 - putkella edellisen käskyn tuote voidaan siirtää seuraavalle käskylle
 
    - for tiedosto in hakemisto/*txt ; do cat $f | grep "koira" ; done 
    
 - for loopin outputin voi ohjata tiedostoon samalla tavalla kuin ilman loopia, joko kaikille tiedostoille yhdessä:
 
     - for tiedosto in hakemisto/*txt ; do cat $f | grep "koira" ; done  > output.txt
     
     - ... tai erikseen: for tiedosto in hakemisto/\*txt ; do cat $f | grep "koira" > $f-output.txt ; done 
 
 
# Tehtävä

- Tee for loop, joka printtaa jokaisen hakemiston conllu-tiedoston nimen ja printtaa sen rivien määrän, ja ohjaa koko tuotos yhteen tiedostoon
